```python
acs_list = [acs_2011, acs_2012,acs_2013,acs_2014,acs_2015,acs_2016,acs_2017,acs_2018]
```

```python
columns = ['white_pop', 'black_pop','asian_pop', 'hispanic_pop','high_school_diploma',
'associates_degree', 'bachelors_degree','masters_degree']
for i, acs in enumerate(acs_list, 1):
    acs['total_pop'] = [1 if x ==0  else x for x in acs['total_pop']]
    acs['housing_units_renter_occupied'] = [1 if x ==0  else x for x in acs['housing_units_renter_occupied']]
    acs['year']= 2010 + i
    for column in columns:
        acs[column] = acs[column]/ acs['total_pop']
    acs['vacant_housing_units_for_rent'] = acs['vacant_housing_units_for_rent']/(acs['vacant_housing_units_for_rent']+acs['housing_units_renter_occupied'])
    acs.drop(columns= ['housing_units_renter_occupied'], inplace =True)
    acs.sort_values(by= ['geo_id'],inplace=True)
```

```python
acs_2018['population_change'] = (acs_2018['total_pop'] - acs_2017['total_pop'])/ acs_2017['total_pop']
acs_2017['population_change'] = (acs_2017['total_pop'] - acs_2016['total_pop'])/ acs_2016['total_pop']
```

```python
acs_2015['total_pop'] = acs_2015['total_pop'].astype(int)
acs_2016['population_change'] = (acs_2016['total_pop'] - acs_2015['total_pop'])/ acs_2015['total_pop']
acs_2015['population_change'] = (acs_2015['total_pop'] - acs_2014['total_pop'])/ acs_2014['total_pop']
acs_2014['population_change'] = (acs_2014['total_pop'] - acs_2013['total_pop'])/ acs_2013['total_pop']
acs_2013['population_change'] = (acs_2013['total_pop'] - acs_2012['total_pop'])/ acs_2012['total_pop']
acs_2012['population_change'] = (acs_2012['total_pop'] - acs_2011['total_pop'])/ acs_2011['total_pop']
```

```python
combined_acs = pd.DataFrame()
for acs in acs_list:
    combined_acs = pd.concat([combined_acs, acs])
```

```python
combined_acs=pd.merge(combined_acs, na, how = 'left', on = 'geo_id')
```
